```python
# Mengimpor library yang dibutuhkan
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

* pandas untuk pengolahan data,
* matplotlib.pyplot dan seaborn untuk visualisasi data.

```python
# Membaca dataset
data = pd.read_csv('/content/sample_data/StudentsPerformance.csv')
```

* Dataset dibaca dari file CSV, kemudian menggunakan data.head() untuk melihat beberapa baris pertama agar kita memahami struktur data.

```python
# Melihat beberapa baris pertama dari dataset
data.head()
```

* data.head() digunakan untuk menampilkan beberapa baris pertama dari dataset yang telah dibaca ke dalam DataFrame data.

```python
# Informasi dataset
data.info()
```

* data.info() menampilkan informasi dasar tentang dataset, seperti tipe data dari tiap kolom dan jumlah data yang ada.

```python
# Statistik deskriptif untuk fitur numerik
data.describe()
```

* data.describe() menampilkan statistik deskriptif (seperti mean, min, max) untuk kolom numerik, yang berguna untuk mendapatkan gambaran awal mengenai distribusi nilai.

```python
# Distribusi nilai ujian berdasarkan gender
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
sns.histplot(data=data, x='math score', hue='gender', multiple='stack', kde=True)
plt.title('Distribusi Nilai Matematika Berdasarkan Gender')

plt.subplot(1, 3, 2)
sns.histplot(data=data, x='reading score', hue='gender', multiple='stack', kde=True)
plt.title('Distribusi Nilai Membaca Berdasarkan Gender')

plt.subplot(1, 3, 3)
sns.histplot(data=data, x='writing score', hue='gender', multiple='stack', kde=True)
plt.title('Distribusi Nilai Menulis Berdasarkan Gender')

plt.tight_layout()
plt.show()
```

Visualisasi ini menampilkan distribusi nilai untuk tiga mata pelajaran berdasarkan gender:

- Histogram menunjukkan perbedaan dalam distribusi nilai antara gender yang berbeda.
- Insight: Misalnya, jika distribusi nilai matematika antara gender berbeda signifikan, itu dapat mengindikasikan perbedaan kemampuan atau kecenderungan belajar di antara mereka.

```python
plt.figure(figsize=(12, 5))
sns.boxplot(x='test preparation course', y='math score', data=data)
plt.title('Pengaruh Kursus Persiapan terhadap Nilai Matematika')
plt.show()
```

Boxplot ini menampilkan nilai matematika siswa yang mengikuti kursus persiapan dibandingkan yang tidak:

- Insight: Median dan rentang nilai mungkin berbeda antara mereka yang mengikuti kursus dan yang tidak. Jika median lebih tinggi pada kelompok yang mengikuti kursus, ini menunjukkan bahwa persiapan membantu performa siswa.

```python
# Korelasi antar nilai ujian
plt.figure(figsize=(8, 6))
sns.heatmap(data[['math score', 'reading score', 'writing score']].corr(), annot=True, cmap='Blues')
plt.title('Korelasi Antar Nilai Ujian')
plt.show()
```

Heatmap ini menunjukkan korelasi antara nilai matematika, membaca, dan menulis:

- Insight: Korelasi tinggi (mendekati 1) antara dua nilai menunjukkan hubungan kuat. Misalnya, korelasi tinggi antara nilai membaca dan menulis menunjukkan bahwa kemampuan literasi mendukung performa pada kedua ujian ini.

```python
# Visualisasi berdasarkan pendidikan orang tua
plt.figure(figsize=(15, 5))
sns.boxplot(x='parental level of education', y='writing score', data=data)
plt.xticks(rotation=45)
plt.title('Pengaruh Pendidikan Orang Tua terhadap Nilai Menulis')
plt.show()
```

Boxplot ini mengamati nilai menulis berdasarkan tingkat pendidikan orang tua:

- Insight: Performa siswa bisa meningkat seiring dengan tingkat pendidikan orang tua yang lebih tinggi, yang menunjukkan pengaruh positif dari latar belakang pendidikan pada nilai ujian.
